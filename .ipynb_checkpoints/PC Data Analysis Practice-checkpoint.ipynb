```python
import np

x = np[200, 235, 300, 330, 360, 400]
y = np[4000, 5500, 9000, 11000, 13000, 16000]

logy = np.log(y)
logx = np.log(x)
logx, logy
```

```python
m, c = np.polyfit(logx, logy, 1)
m, c
```

```python
np.exp(c)
```

```python
import matplotlib.pyplot as plt

X = np.linspace(logx[0], logx[-1], 1000)
plt.figure(figsize=(16,8))
plt.scatter(logx, logy)
plt.plot(X, m*X + c)
```
